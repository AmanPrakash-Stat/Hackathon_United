```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading csv files
t1 = pd.read_csv(r"E:\united_project\callsf0d4f5a.csv")
t2 = pd.read_csv(r"E:\united_project\customers2afd6ea.csv")
t3 = pd.read_csv(r"E:\united_project\reason18315ff.csv")
t4 = pd.read_csv(r"E:\united_project\sentiment_statisticscc1e57a.csv")
```

```python
# Combining the tables in one table using customer_id and call_id.
merge_table = pd.merge(t1,t2, on = "customer_id",how="inner")
merge_table = pd.merge(merge_table,t4, on = "call_id",how="inner")
merge_table = pd.merge(merge_table,t3, on = "call_id",how="left")
```

```python
# to find call duration:
merge_table['call_start_datetime'] = pd.to_datetime(merge_table['call_start_datetime'])
merge_table['call_end_datetime'] = pd.to_datetime(merge_table['call_end_datetime'])
merge_table["call_duration"] = merge_table['call_end_datetime']-merge_table['call_start_datetime']
merge_table["call_duration"] = merge_table['call_duration'].dt.total_seconds() // 60
```

### Multiple Regression Model:

```python
df = df.dropna()
```

```python
df.shape
```

```python
y = df["call_duration"]
x = df[["agent_tone",'customer_tone', 'average_sentiment',
       'silence_percent_average']]
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
print(lr.score(x_test,y_test))
print(lr.score(x_train,y_train))
print(lr.coef_)
```
